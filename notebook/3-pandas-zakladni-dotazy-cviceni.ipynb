```python
import pandas
```

Načti soubor `jmena.csv` tak, aby Pandas vyrobil číselný index.

```python
jmena = pandas.read_csv('jmena.csv')
jmena
```

Vypiš všechny řádky se jmény, jejichž nositelé mají průměrný věk vyšší než 60.

```python
jmena[jmena["věk"] > 60]
```

Vypiš pouze jména z těch řádků, kde četnost je mezi 80 000 a 100 000.

```python
# Varianta 1
jmena[(jmena["četnost"] > 80_000) & (jmena["četnost"] < 100_000)]["jméno"]
```

```python
# Varianta 2
jmena[jmena["četnost"].between(80_000, 100_000)]["jméno"]
```

Vypiš jména a četnost pro jména se slovanským nebo hebrejským původem.

```python
jmena_sl_he = jmena[jmena["původ"].isin(["hebrejský", "slovanský"])]
jmena_sl_he[["jméno", "četnost"]]
```

 Kolik takových jmen je?

```python
jmena_sl_he.count()
```

```python
len(jmena_sl_he)
```

```python
jmena_sl_he.shape
```

Vypiš všechna jména, která mají svátek první 3 dny v prosinci.

```python
jmena[jmena["svátek"].isin(["1.12", "2.12", "3.12"])]
```

```python
jmena[jmena["svátek"].isin(["1.12", "2.12", "3.12"])]["jméno"]
```
